```python
import sys 
############################################################################################################
#######################   ACCIONES DENTRO DE ARCHIVOS
############################################################################################################
def leer_archivo(direccion_r):
    try:
        path = f"./{direccion_r}"
        abrir_archivo = open(path)
        lector = abrir_archivo.readlines()
        abrir_archivo.close()
    except Exception as Error:
        print("Error en lectura de archivo.")
    return lector

def ingresar_data(direccion_a, 
                  data):
    try:
        path = f"./{direccion_a}"
        editar_archivo = open(path,'a')
        editar_archivo.write(f"\n{data}")
        editar_archivo.close()
    except Exception as Error:
        print("No se pudo modificar su archivo.")
        
def eliminar_data(direccion_d, 
                  dato):
    lista_data = list(leer_archivo(direccion_d))
    for i in lista_data:
        if i == dato:
            del lista_data[lista_data.index(i):lista_data.index(i)+6]
    return lista_data
    
def modificar_data(direccion_m, 
                   dato_registrado, 
                   dato_nuevo):
    lista_data_mod = list(leer_archivo(direccion_m))
    for i in lista_data_mod:
        if i == f"{dato_registrado}\n":
            lista_data_mod[lista_data_mod.index(i)] = f"{dato_nuevo}\n"
            return lista_data_mod
    
def buscar_data(direccion_b, 
                elemento_buscar):
    lista_data_buscar = list(leer_archivo(direccion_b))
    for i in lista_data_buscar:
        if i == elemento_buscar:
            return lista_data_buscar[lista_data_buscar.index(i):lista_data_buscar.index(i)+6]

def ingresar_nueva_data(direccion_nd):
    try:
        path = f"./{direccion_nd}"
        editar_archivo = open(path,'r+')
        lista = str(eliminar_data("instrumentos.txt", "Guitarra\n"))
        for i in lista:
            editar_archivo.write(i)
        editar_archivo.close()
    except Exception as Error:
        print("No se pudo eliminar el artículo de su archivo.")
        
############################################################################################################
#######################   INSTRUMENTOS
############################################################################################################
def crear_instrumento():
    tipo_ci = input("Ingrese el tipo de instrumento que desee crear: ")
    marca_ci = input("Ingrese la marca del instrumento: ")
    modelo_ci = input("Ingrese la modelo del instrumento: ")
    color_ci = input("Ingrese la color del instrumento: ")
    precio_ci = input("Ingrese el precio del instrumento: ")
    peso_ci = input("Ingrese la peso del instrumento: ")
    ingresar_data("instrumentos.txt", f"{tipo_ci}\nMarca: {marca_ci}\nModelo: {modelo_ci}\nColor: {color_ci}\nPrecio: {precio_ci}\nPeso: {peso_ci}")

def eliminar_instrumento():
    instrumento = input("Ingrese el instrumento a ser eliminado: ")
    lista = eliminar_data("instrumentos.txt", f"{instrumento}\n")
    print(lista)
    return lista

def buscar_instrumento():
    instrumento_buscar = input("Ingrese el instrumento que desea buscar: ")
    caracteristicas = buscar_data("instrumentos.txt", f"{instrumento_buscar}\n")
    print(caracteristicas)
    return caracteristicas

def mostrar_todos_instrumentos():
    instrumentos = leer_archivo("instrumentos.txt")
    print(instrumentos)
    return instrumentos

############################################################################################################
#######################   GENEROS MUSICALES
############################################################################################################
def crear_genero():
    genero = input("Ingrese el genero musical que desea crear: ")
    fecha_creacion = input("Ingrese la fecha de creacion: ")
    pioneros = input("Banda pionera: ")
    banda_preferencia = input("Banda preferida: ")
    escala_utilizada = input("Escala comunmente usada: ")
    instrumentos = input("Ingrese los instrumentos usados: ")
    ingresar_data("archivo_genero.txt", f"{genero}\nFecha de creacion: {fecha_creacion}\nBanda pionera: {pioneros}\nTu banda preferida: {banda_preferencia}\nEscala utilizada: {escala_utilizada}\nInstrumentos usados: {instrumentos}")
    
def eliminar_genero():
    genero = input("Ingrese el genero a ser eliminado: ")
    lista = eliminar_data("archivo_genero.txt", f"{genero}\n")
    print(lista)
    return lista

def buscar_genero():
    genero_buscar = input("Ingrese el instrumento que desea buscar: ")
    caracteristicas = buscar_data("archivo_genero.txt", f"{genero_buscar}\n")
    print(caracteristicas)
    return caracteristicas

def mostrar_todos_generos():
    generos = leer_archivo("archivo_genero.txt")
    print(generos)
    return generos

############################################################################################################
#######################   MENU
############################################################################################################
def menu():
    print("############-lml-Bienvenido a Music Land-lml-###########")
    print("En esta aplicación podras ingresar tus gustos musicales.")
    print("1. Generos creados")
    print("2. Instrumentos creados")
    print("3. Opciones en generos musicales")
    print("4. Opciones en instrumentos musicales")
    print("9. SALIR")
    opcion = input("Ingrese una opción: ")
    if opcion == "1":
        mostrar_todos_generos()
    elif opcion == "2":
        mostrar_todos_instrumentos()
    elif opcion == "3":
        acciones_generos()
    elif opcion == "4":
        acciones_instrumentos()
    elif opcion == "9":
        sys.exit()
    else:
        print("**Ingrese una opción correcta**")
        
def acciones_generos():
    print("1. Ingresar un genero musical")
    print("2. Eliminar un genero musical")
    print("3. Buscar un genero musical")
    print("0. SALIR")
    opcion_g = input("Ingrese una opción: ")
    if opcion_g == "1":
        crear_genero()
    elif opcion_g == "2":
        eliminar_genero()
    elif opcion_g == "3":
        buscar_genero()
    elif opcion_g == "0":
        breakpoint
    else:
        print("**Ingrese una opción correcta**")
    
def acciones_instrumentos():
    print("1. Ingresar un instrumento musical")
    print("2. Eliminar un instrumento musical")
    print("3. Buscar un instrumento musical")
    print("0. SALIR")
    opcion_i = input("Ingrese una opción: ")
    if opcion_i == "1":
        crear_instrumento()
    elif opcion_i == "2":
        eliminar_instrumento()
    elif opcion_i == "3":
        buscar_instrumento()
    elif opcion_i == "0":
        breakpoint
    else:
        print("**Ingrese una opción correcta**")

#menu()
        
#leer_archivo("instrumentos.txt")
#ingresar_data("archivo_genero.txt", "Bajo\n\tMarca: Jackson\n\tModelo: JS12\n\tColor: Negro\n\tPrecio: 850\n\tPeso: 25.0")
#buscar_data("instrumentos.txt", "Guitarra\n")
#eliminar_data("instrumentos.txt", "Guitarra\n")
#modificar_data("instrumentos.txt", "Piano", "Ukelele")
ingresar_nueva_data("instrumentos.txt")
#crear_instrumento()
#eliminar_instrumento()
#buscar_instrumento()
```
