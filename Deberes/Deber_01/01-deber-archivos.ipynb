```python
def leer_archivo(direccion_r):
    try:
        path = f"./{direccion_r}"
        abrir_archivo = open(path)
        lector = abrir_archivo.readlines()
        abrir_archivo.close()
    except Exception as Error:
        print("Error en lectura de archivo.")
    return lector

def ingresar_data(direccion_a, 
                  data):
    try:
        path = f"./{direccion_a}"
        editar_archivo = open(path,'a')
        editar_archivo.write(f"\n{data}")
        editar_archivo.close()
    except Exception as Error:
        print("No se pudo modificar su archivo.")
        
def eliminar_data(direccion_d, 
                  dato):
    lista_data = list(leer_archivo(direccion_d))
    for i in lista_data:
        if i == dato:
            del lista_data[lista_data.index(i):lista_data.index(i)+6]
    return lista_data
    
def modificar_data(direccion_m, 
                   dato_registrado, 
                   dato_nuevo):
    lista_data_mod = list(leer_archivo(direccion_m))
    for i in lista_data_mod:
        if i == f"{dato_registrado}\n":
            lista_data_mod[lista_data_mod.index(i)] = f"{dato_nuevo}\n"
            return lista_data_mod
    
def buscar_data(direccion_b, 
                elemento_buscar):
    lista_data_buscar = list(leer_archivo(direccion_b))
    for i in lista_data_buscar:
        if i == elemento_buscar:
            return lista_data_buscar[lista_data_buscar.index(i):lista_data_buscar.index(i)+6]

def ingresar_nueva_data(direccion_nd):
    try:
        path = f"./{direccion_nd}"
        editar_archivo = open(path,'r+')
        lista = str(eliminar_data("instrumentos.txt", "Guitarra\n"))
        for i in lista:
        #print(type(lista))
            editar_archivo.write(lista)
        editar_archivo.close()
    except Exception as Error:
        print("No se pudo eliminar el art√≠culo de su archivo.")
        
def crear_instrumento():
    tipo_ci = input("Ingrese el tipo de instrumento que desee crear: ")
    marca_ci = input("Ingrese la marca del instrumento: ")
    modelo_ci = input("Ingrese la modelo del instrumento: ")
    color_ci = input("Ingrese la color del instrumento: ")
    precio_ci = input("Ingrese la precio del instrumento: ")
    peso_ci = input("Ingrese la peso del instrumento: ")
    ingresar_data("instrumentos.txt", f"{tipo_ci}\n\tMarca: {marca_ci}\n\tModelo: {modelo_ci}\n\tColor: {color_ci}\n\tPrecio: {precio_ci}\n\tPeso: {peso_ci}")
    
            
leer_archivo("instrumentos.txt")
#ingresar_data("instrumentos.txt", "Bajo\n\tMarca: Jackson\n\tModelo: JS12\n\tColor: Negro\n\tPrecio: 850\n\tPeso: 25.0")
#buscar_data("instrumentos.txt", "Guitarra\n")
#eliminar_data("instrumentos.txt", "Guitarra\n")
#modificar_data("instrumentos.txt", "Piano", "Ukelele")
#ingresar_nueva_data("instrumentos.txt")
#crear_instrumento()
```
